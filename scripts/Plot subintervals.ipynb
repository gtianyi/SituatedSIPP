```python
import subintervals
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
subinterval_set = subintervals.SubIntervals(0.0, 100.0)
subinterval_set.add(20.0, 50.0, 10)
subinterval_set.add(60.0, 80.0, 20)
plt.figure()
subinterval_set.plot()
plt.show()
```

```python
subinterval_set = subintervals.SubIntervals(0.0, 100.0)
subinterval_set.add(20.0, 50.0, 40)
subinterval_set.add(60.0, 80.0, 10)
subinterval_set.add(70.0, 100.0, 20)

plt.figure()
subinterval_set.plot()
plt.show()
```

```python
subinterval_set = subintervals.SubIntervals(0.0, 100.0)
subinterval_set.add(20.0, 50.0, 60)
subinterval_set.add(60.0, 80.0, 10)
plt.figure()
subinterval_set.plot()
plt.show()
```

```python
subinterval_set = subintervals.SubIntervals(0.0, 100.0)

subinterval_set.add(0, 51, 10)
subinterval_set.add(40, 60, 10.5)
subinterval_set.add(49, 100, 9.5)
plt.figure()
subinterval_set.plot()
plt.title("run")
plt.show()

subinterval_set = subintervals.SubIntervals(0.0, 100.0)

subinterval_set.add(0, 51, 10)
subinterval_set.add(49, 100, 10)
plt.figure()
subinterval_set.plot()
plt.title("minimal run")
plt.show()
```
